```python
# A NOTE ON **************T-TESTS!

T values are a metric of difference between __two groups__ of data.
A t-value of "3" means that the values of group 1 are 3 times as different from group 2, as well as within each other

LARGER THE T-VALUE, LARGER THE DIFFERENCE. SMALLER THE T-VALUE, THE MORE SIMILARITY BETWEEN GROUPS

t-value = abs((meangroupx - meangroupy)) / (sqrt(stdx^2/len(x) + stdy^2/len(y)))

Compare t-value with "t-chart", which shows t-values for comparison of significance
If experimental t-value > t-chart value, then statistically significant difference between groups 
(i.e., you would reject null hypothesis in such case)
```
