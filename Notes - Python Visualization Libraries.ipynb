```python
# How to do a correlation heatmap!

# First, create a correlation variable
# The df will contain all and only the datacolumns you want to analyze
corr = dfheat.corr()

# Always good to have as a supplement
print(corr)

# Generating a mask for the upper triangle of the correlation heatmap
# This is purely optional and may not be necessary...
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True


# vmax and vmin are for the limits of the correlation (usually -1 to 1)
# annot copies the numerical values into the boxes
plt.figure(figsize=(20,10))
ax = sns.heatmap(corr, mask=mask, vmax=0.5, 
                 xticklabels=dfheat.columns, yticklabels=dfheat.columns, annot=True)
```
