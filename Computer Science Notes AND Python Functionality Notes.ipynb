_Calculating binary to decimal:_

binary = 10010

## Functions

Parameters of Python functions:

- Functions _can_ be defined within another function ("Function Nesting").
- Functions do _not_ require a "return" statement. Functions can have zero or more return statements.
- If and for statements do not _always_ require indentation (e.g., they can be part of a one-line list comprehension).
- Keyword arguments are _not_ the same as default parameter values. This is more of an academic distinction, but correct terminology is nonetheless important.
    - for a function `def func(a=1, b=2, c=3): pass`: 
        - a, b, and c are the default parameter/argument values. If you did func(), it would readily assume the default values of 1, 2, and 3, respectively.
        - Conversely, keyword arguments are _called_ - they are not the default. This allows for out-of-order calls too, like func(c=100, a=12, b=20). If "b" wasn't passed as an argument, again, it would assume the default value of b=2.
- Breaks in a function are used to stop and exit from executing _only_ the enclosed loop in which the "break" resides.
    
### Global vs. Local Differences

- __Global__ means outside of a function, i.e., applies globally. __Local__ means defined and used only in a specific context, such as only within a function.
- For example, "global built-in functions" are list, sum, and other basic Python built-in functionality that can be used both inside and outside functions.
    - Global built-in functions _can_ be redefined (e.g., `list = 'alfie'` making `list(0,1,2,3)` broken), but that is ill-advised.
    

- The __LEGB Rule__ identifies the order in which Python searches a variable for use:
    - __(L)__ _Local_: Inside a function or class. If the variable is not found there, it will look to 
    - __(E)__ _Enclosed_: Outside portions of the function (not local to the specific indented level of the function).
    - __(G)__ _Global_: If not in a function, will look to global variables. That's why list = alfie changes list before
    - __(B)__ _Built-in_: If not defined globally, will look to built-in functionality for variable's purpose.
    <br><br>
    - Storing local variables are __faster__ to lookup than global variables. _Why?_ The bytecode uses a `STORE_FAST` action instead of the global variable's `STORE_NAME` action. Why this is the case just comes down to convention (based on my research).

## Boolean Operators

### True vs. False

Things that output ___False___:
- 0 
- '' (empty string)
- False


Things that output ___True___:
- Everything else!
    - '0' (all non-empty strings, including strings of the numeral zero)
    - ' ' (string with space are non-empty strings)
    - -1 (all numbers except zero, including negative numbers)
    
### Specific Boolean Points

- "is" and "==" are two different things! "==" is a boolean test for _equivalence_ while "is" is a stricter test for _identity_.

```python
list1 = []
list2 = []
list3 = list1

list1 == list2 # outputs TRUE
list1 is list2 # outputs FALSE
list1 is list1 # outputs TRUE
```
