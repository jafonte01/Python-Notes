## A.1. Binary: What it is and Why we Use it

Instead of "thresholds", we use the abstraction of "complete electrical signal", or "1", and "zero electrical signal", or "0". Using these ___binary___ values, we can use binary combinations to represent data, and set locations of where to store such data. 0's and 1's are bits:


__Bit__: A single unit of information. It is a truncated word for "binary digit"; however, bits can be represented either in binary (0b------) or hexadecimal (0x-------).

### Calculating Binary to Decimal

#### Converting Binary to Decimal

Each digit of binary represents an exponent of base 2. Five digits of binary represent up to 2<sup>4</sup> (the fifth digit is 2<sup>0</sup>). Add up all the existing places with the value "1" to get the decimal value.


__Example:__


For the binary 10010 (= 18), do:


(2<sup>4</sup> x 1) + (2<sup>3</sup> x 0) + (2<sup>2</sup> x 0) + (2<sup>1</sup> x 1) + (2<sup>0</sup> x 0) = 18 

#### Converting Decimal to Binary

To convert decimal to binary:


1. Determine the exponent for 2<sup>x</sup> that yields the square value immediately higher than the decimal value. Use x digits for representing the decimal: we do _not_ include 2<sup>x</sup> (-1) but we _do_ include 2<sup>0</sup> (+1). 
    - If the decimal is a perfect square, like 32 (2<sup>5</sup>), then use (x + 1) digits, i.e., 100000. 
2. From the largest exponent of (x-1) to the smallest (0), use a 1 if including the exponent yields a decimal value less than or equivalent to the target decimal value. If the binary yields an equivalent decimal value, make the remaining digits zeroes.


__Example:__


Let's convert the decimal ___23___ to binary.


The first perfect square of two that is immediately larger than 23 is 32, or 2<sup>5</sup>; the next largest perfect square, 2<sup>4</sup>, or 16, is too small of a value to represent 23. According to step 1 above, we need ((5) - 1 + 1) = 5 binary digits to represent 23.


For the second step, we start with the largest exponent in our binary, or ((5) - 1) = 4, and work our way 1 by 1 to the target decimal value of 23.


(2<sup>4</sup> x 1) = 16. Our starting exponent is a 1 (it always is in conversions, see the "Hexadecimal" subsection as to why leading zeroes exist at all).


Do we need a 2<sup>3</sup>? No, because 16 + 2<sup>3</sup> = 24, i.e., over our target value. This binary value is set to 0.


Next, do we need a 2<sup>2</sup>? Yes, because 4 brings us closer to 23 but does not bring us over this value: Now, the binary value is (16 + 4) = 20. Make this place a 1.


For the final two values, 2<sup>1</sup> and 2<sup>0</sup>, we need both of those too to get 23. Both of these places get a 1.


__The resulting binary value for 23 is: 10111.


__NOTE!__ Any odd decimal value yields a binary number ending in 1, since 2<sup>0</sup> is the only way to produce an odd value. Conversely, if the decimal value is even, we know the last binary number is a 0.

## A.2. Hexadecimal and Why We Use It

__Question:__ If binary works so well for representing information, why do we need another base system for the same representation?


__Answer:__ Historically, computers used four-bit architectures, making binary representations as sets of four:


50 = 110010, we would write: 0011 0010. 
- This is why we have leading zeroes in data!


For much longer strings of binary numbers, the manual calculation becomes cumbersome. __HOWEVER__, four digit binary values yield 16 (2<sup>4</sup>) combinations, which can be consolidated into one single number, a ___hexadecimal___.


### Calculating Decimal to Hexadecimal and Vice-Versa

The methods of converting decimal to hexadecimal and hexadecimal to decimal are the same as binary to decimal conversion, except we use base 16 instead of base 2. 


The only difference is that we take it one step further and determine what the multiplier is for every digit place, instead of slapping a "1" on it. We do this in the same manner: which multiplying number brings us the closest to the target decimal value without bringing us over the value.


__Example:__ For the number 50, we use need 2 digits, because 16<sup>2</sup> is immediately larger than 50 and 16<sup>1</sup> is immediately smaller than 50.


For the 16<sup>1</sup> place, we find that the multiplying number without bringing us over 50 is 3. We then use the multiplying number in the 16<sup>0</sup> place to fill the remainder to get us to 50:


(16<sup>1</sup> x 3) + (16<sup>1</sup> x 2) = 50. __The hexadecimal number for the decimal value 50 is "32".__


__Another example:__ What is the hexadecimal number for the decimal value 45?


Still two digits, (16<sup>1</sup> x 2) + (16<sup>1</sup> x 13) = 45. But we can't write 2(13). ___This is where letters come in.___ We use A for a hexadecimal value of "10", B for "11", up to F for "15". Therefore, __the hexadecimal number for the decimal value 45 is 2C.__

### Getting back to the Relation of Hexadecimal and Binary

So why did I use the value "50" in both the introductory section _and_ the calculation section? Let's take a look again:


Decimal value:......... 50 <br>
Binary value:.......... 0011 0010 <br>
Hexadecimal value:..... 32


What a coincidence! The first set of four binary numbers yields a decimal value of 3 and the second set yields a decimal value of 2, the same numbers as the hexadecimal value. Again, because four binary digits can represent sixteen different combinations, all we did was compress it down to a base that can represent all sixteen combinations (base ___16___). This is much easier to interpret than a long string of binary numbers.

# B. Compression

## B.1. The Definition and Lossless vs. Lossy

- Two types of compression: ___lossless___ and ___lossy___.
    - Lossless means no information is lost during the compression. The original file can be reconstructed ("unzipped") from the lossless compressed file.
        - An example of this is compressing a text file by replacing strings of repeated characters with single symbols (e.g., a diamond for every "th" in the file, or a dollar sign for ever "does not" in the file). 
            - This example compression technique is called ___byte pair encoding___.
            - Most files do contain _some_ repeated sequences of bits, making lossless compression frequently possible.
    - Lossy means the opposite: information is lost during the compression that cannot be retrieved. An example is a reduced quality version of an image for saving space.

***

## B.2. Lossless Compression: RLE

Images are compressed differently from text (obviously). The main algorithm for simple image compression is ___Run-Length Encoding (RLE) Compression___. An original image is represented by a ___bitmap___, which every pixels is assigned its own bit value to represent color or no color (oversimplified, but you get the point). With RLE, the string of consecutive 0's and consecutive 1's is compressed down to a single number representing the length of the consecutive number. <br><br>
For example:
<br><br>
0011100 <br>
0001000 <br>
0011100 <br>
<br><br>
becomes<br><br>
232<br>
313<br>
232<br>
<br><br>
Every pixel row has now been condensed via RLE from 7 bits long down to 3 bits. RLE ___Decompression___ is the opposite: it takes the numerals and expands them into a bitmap of the image.


__Note:__ RLE only works if consecutive strings of values exist in a bitmap. If every pixel of an image is a different color or different shade - even if the shades are similar but not exact - RLE won't work. As such, RLE is popular with computers that have limited color palettes. Therefore, RLE is a simplistic example of image compression but is not widely used today. Fax machines and scanned black and white letters have RLE compression, and RLE compression is often a final step in much larger and more sophisticated image compression algorithms.

## B.3. Lossless Compression: Bit Compression via Huffman Coding Algorithm

If computers interpret any kind of data as bits, then we can apply compression algorithms on the _bits_ representing whatever data we want. Bit compression is done with the following algorithm:


### Huffman Coding Algorithm

We set bit identifiers to represent every unique character in text. The algorithm makes shorter identifiers for those characters that appear the most frequently in the text, which saves space.


For example, if we want to assign bit values to the DNA string `AGCTTTTCATTCT`, we need 2 bits (to identify four total unique things). Therefore, we could make <br><br>

| DNA Base | Bit Representation
| :---: | :----:|
| A | 00 |
| C | 01 |
| G | 10 |
| T | 11 |


Substituting the bits for the DNA bases, we get the bit code: __00100111111111010011110111__, 26 bits long. BUT, the Huffman Coding Algorithm sees that "T" is represented the most, and so it can compress the bit code even more by making the bit identifier for T shorter (while maintaining each DNA base to be readily identifiable with a bit encoding), and in turn, making the least frequent bases have a larger bit representation:


| DNA Base | Bit Representation
| :---: | :----:|
| A | 010 |
| C | 00 |
| G | 011 |
| T | 1 |


This gives us the encoding of: __0100110011110001011001__, 22 bits long.


Huffman Coding is often used in conjunction with another compression algorithm (e.g., after RLE) by identifying the most frequent elements and shortening that at the expense of enlarging rarer elements.

## B.4. Lossy Compression

Lossy compression is often associated with image compression because it can reduce the size of the file _significantly_ without a human being able to tell the difference. No one likes to permanently lose data, but if you can't tell the difference, who cares?


__Example:__ [this photo of a cat](https://cdn.kastatic.org/ka-perseus-images/249d175f16a1081aaeb64636b4a99ad988871b1d.jpg) can be reduced to __HALF__ the file size without losing any discernable quality, as shown in [the compressed version of that photo](https://cdn.kastatic.org/ka-perseus-images/acf4c90e754c7e8b2f4430908459fb6fca9e8fee.jpg) (yes, I swear they are two different photos!). In this example, the photo is compressed by averaging the colors in the photo, which is much harder to discern by humans than the brightness of areas of the photo. 


The compression technique described above is ___chroma subsampling___, which optimized by the "discrete cosine transform algorithm" (beyond scope of notes). Often, chroma subsampling can be done to reduce an image's quality without any discernable difference down to _10%_ of its original file size before a difference in quality can be ascertained. Thus, not surprisingly, chroma subsampling is a quite popular compression method for images and videos, especially when using the web to transmit media data over internet bandwidth.  


The same idea of dropping the indiscernable portions of a file is also done with audio files; audio files can be compressed by dropping pitches that are outside a human's hearing range.

## A.1. Introduction

Like a transistor's input and output, a computer's interpretability of binary is based on active electrical signals (1) and the absence thereof (0). In a wire, such as a LAN connection or ethernet cable connected to an internet router, electrical signals are sent in a Morse code fashion - sending signals and absence of signals in quick succession to represent binary.


So how do we know that a computer sent 5 consecutive 0's, since there is an absence of signal? The answer lies in the ___bit rate___ of a data transfer: a defined amount of time for accepting a signal. Bit rate is measured in bits per second, "bps" For example, if the bit rate is 1 bps, then after five seconds of no signal, the transferred binary would be 00000.

## B.1. The Basic Architecture of the Internet: Address Protocols

A ___protocol___ is a set of uniform rules and standards adopted by multiple computers in a network. Each computer is given a unique protocol address, i.e., its __Internet Protocol (IP)__ address. Under an Internet Protocol, a unique address sends a request for information to a destination address. As part of the request, the recipient also receives the requestor's IP address, so that it knows where to send the requested information to.

### How to Read an IP Address

IP addresses are now written in four sections, called __IPv4__. Take the following example IP address:


174.129.14.120


Each section is 8 bits long (i.e., an IP address is 32 bits long total). Because they are 8 bits long, the number for each section can only be from 0 to 255 (2<sup>8</sup> unique values).


The numbers are hierarchical: the first number refers to the network and the last number refers to an individual node in the network. Specifically, the first two numbers refer to more global locale information: the country's network or regional network (more network-based than geographic-based, however). The third number is a sub-network designation, and the final number refers to the specific device. For example, "174.129" may be the Comcast Network, and "14.120" is your home network. Another example showing a different configuration is "174.129" is the University of Michigan's computer network, "14" for its Medical School subnetwork, and "120" for a specific lab computer.


As of 2020, we are currently in a multi-year transition to IPv6, which has 6 sections each 16 bits long, due to a lack of unique addresses.

## B.2. The Domain Name System

The Domain Name System (DNS) associates names, like a web URL, with the corresponding IP address. The system is not one server: it is multiple DNS servers. This is an example of how it works:


1. A computer with an IP address makes a request to a DNS server for the IP address of a specific name. It can either send back the IP address associated with the name so that the computer goes to the appropriate website, but it is possible that specific DNS server does not have that information (since the Internet is so big and its connections changing constantly). In such a case: 
2. This DNS server would refer to another DNS server if the address for the requested name-IP address association.
3. Another server that does have the IP address for the requested name transfers the information to the initially requested server, which saves the IP address - requested name association for future requests. 
4. The initial DNS server transfers the associated IP address to the requesting computer, which then goes to the appropriate location.


In short, it is a series of servers that constantly update each other, which is a more feasible way than someone making sure every DNS server is constantly updated, or the idea of a single giant DNS server answering IP-name requests for the entirety of the Internet. DNS servers are organized in a zone-based hierarchy. For example: "blog.gardeninghow.org" has .org as the "Top Level Domain", the gardeninghow as the "Second-Level Domain", and blog as the "Third-Level Domain".


This organization, along with IP addresses, are designed to scale to accommodate a growing Internet.

# C. Reliable Data Transfer: TCP/IP

How do we know the data transferred from computer to computer is, in fact, sent and received by the destination computer? In a more general sense, what can computers do to address the [Two Generals' Problem](https://en.wikipedia.org/wiki/Two_Generals%27_Problem)? This section addresses the internet protocols in place to ensure reliable data transfer.

***

## C.1. Why Reliability is an Issue for the Internet

### Introduction

Data requests are not directly made to a target website, and the target website's response is sent directly back to you; it is not like there is a dedicated line between your terminal and the target website's server. The internet would be immediately overloaded if that were the case. Instead, requests are made much more indirectly...

### How Data is Transferred on the Internet

Data is transferred as information __IP packets___, taking a variable path to the destination server, depending on a particular path's congestion, location, general speed of the wires, and the size of the packet. At every connection, the packet is re-routed via a __router__ to get it closer to the destination.


Often, information packets are too large to send at once, especially media data. Sending one picture for example would require the internet to break up the billions of bits into multiple smaller packets, which do not arrive at the destination all at the same time (and may not even arrive in order). The specifics of how this works is described in section C.2. below. How the Internet ensures that all packets are sent is described in section C.3. below.


### Reliability of Internet Transfer Requires Redundancies

Having multiple paths for packets to be transferred seems like a lot of redundancy in the system. However, there are two main positives of these redundancies: 
1. Redundancies make the system ___fault tolerant___, which is ability ("tolerance") of a system to continue executing and completing a task even if something goes wrong. 
    - In the case of the Internet as the "system", if a specific path is broken, unresponsive, or otherwise is unable to transfer the desired packet).
2. Redundancies make the system more flexible in scaling and accommodating load. This is essential for the Internet.

***

## C.2. Internet Routing Protocol - The Specifics of Data Transfer

__Question:__ How does the Internet know which path to take?
- __Answer:__ __Internet Routing Protocol ("IP")__ chooses the ___cheapest available path___ for _each_ packet, where "cheapest" means time sent and in accordance with ISP agreement/relationships.


### Step-by-Step Example
1. Computer with IP address sends information packet to nearest router. The router reads the desired destination IP address, and it forwards the packet via the router's ___forwarding table___, which has IP address prefixes ordered by hierarchical preference (again, based on network agreement, distance, size of path, etc.). 
2. Based on the first listed IP address in the router's forwarding table that matches the destination IP address of the packet, it forwards it to another router, which does the same analysis in where to re-route the packet.
3. This continues until the packet reaches the destination IP address, which is often either a personal computer or server.

***

## C.3. Transmission Control Protocol (TCP)

### Problems with the "Multiple Packet" Framework of Sending Data

As stated in C.1., there are a number of reasons why packets may not reach the destination:
- Problems in the physical layer (broken wire).
- Problems with the forwarding table of a router(s).
- Packets arriving out of order.
- Packets being misattrbuted to the wrong request, which can happen if one computer makes multiple requests to the same destination (thus having the same source IP address and destination IP address).

### TCP - An Overview

To address this, we have __Transmission Control Protocol (TCP)___ alongside the aforementioned Internet \[Routing\] Protocol. It is commonly known together as TCP/IP.


- TCP is the "guaranteed mail service" that ensures reliability of the transfer of _all_ data packets for a specific request. 
- If even one packet is not sent, TCP rejects the delivery and requires the sender to send the request (or response) again - again checking all packets to be sent before accepting the transferred data.

### TCP - Step-by-Step Analysis

1. _Establishing a Connection with the Next Node_: TCP uses a "__Three-Way Handshake__" to confirm between nodes that a connection is established:
    - Sending Node sends a "Synchronize?" ("SYN") message to the receiving node.
    - Receiving node sends back an "Acknowledge!" ("ACK SYN") message to the sending node.
    - The sending node sends back an ACK message to let the receiving node know it recognizes its acknowledgement of the original synchronize request.
   
   
2. _Sending Data in Packets_: One by one, the sending node sends the data packet, which includes the data along with TCP information, such as the sequence # of the data packet (as to ascertain the original order of the packets). One by one, the receiving node sends an ACK back for each received packet.
    - If no ACK message is sent back after a certain period of time, the sending node will automatically send the same packet again.


3. _Notifying Termination of Transmission_: When the transmission is complete (or is cancelled or otherwise terminated), the sending node sends a "Finish" ("FIN") message. The receiving node will send an ACK FIN message back, with a final confirming ACK message from the sending node to the receiving node. Once this happens, the connection is closed.

# D. Data Security During Transfer

For such an open interconnected system where everyone uses the same wires for data transfer, there must be a security measure in place if private and sensitive data is also to be transferred over these lines.

## D.1. Basics of Encryption

__Encryption__ is a method of converting information into different information by a specific algorithm (e.g., "every letter now becomes the letter three after the original letter in the alphabet"). Likewise, __decryption__ is translating the encrypted information back into the original information.

### Encryption Keys

The specific algorithm is called the ___key___, such as "alphabet + 3". In reality, every letter of the information package would be encrypted in a different way for it to be more secure (first letter + 3, second letter + 7, etc.). The sender and recipient must both have the key for encryption to work.

***

## D.2. Types of Encryption

An "alphabet + 3" type of encryption is known as ___symmetric encryption___, where the encryption method and decryption method are the same. Computers can crack these methods quickly (for an alphabet + 3, one would have to just shift all the letters at most 26 times...). 


There are many kinds of encryption, including shift key encryption (a code word ("cipher") is used to identify how to decrypt, and the AES-128, the federal government's encryption where each of the 128 bits is encrypted with 10 mathematical operations. Most computer's use:

### Public Key Encryption a/k/a Asymmetric Encryption

*Now*, computers use ___asymmetric keys___ for better security, a public key that both the sender and recipient have to identify the encryption, and an unseen private key used by the recipient computer for decryption. 
- The public key is a one-way encryption; one simply can't "undo" it. 


__Public key cryptography is the foundation of all secure messaging on the internet.__


### Public Key Encryption Step-by-Step

1. The sending node ("client") generates a public key for encryption and a private key for encryption. Meanwhile, the receiving node ("server") also generates a public key and private key.
2. The two public keys are exchanged between the nodes. The private keys are never exchanged.
3. The sending node encrypts the data with the receiving node's public key.
4. The sending node sends the encrypted data to the receiving node.
5. The receiving node uses its own private key to decrypt the data.
6. If the receiving node needs to send encrypted data back to the sending node, it will use the sending node's public key to encrypt, where the sending node will decrypt it with its own private key once received.

***

## D.3. Transport Layer Security (TLS)

TLS (formerly known as "Secured Sockets Layer" ("SSL")) takes advantage of both symmetric encryption and public key encryption, along with miscellaneous additional features such as authentication and message tampering detection.
- The additional security layers addes some latency in data transmission, but this negative is greatly outweighed by the security.
    - Public key encryption, however, takes _significantly_ more time than symmetric encryption due to taking more time to performing mathematical operations to create uncrackable keys.

### TLS Step-By-Step

1. Three-way handshake is established between sending ("client") and receiving ("server") nodes.
2. After three-way handshake, the client sends a message requesting that a TLS connection is made instead of a standard insecure connection.
3. The server sends back a __Digital Certificate__ containing its public key. The client verifies the certificate as valid before further action.
    - If the server does not support TLS Protocol, it will abort the connection.
    - If the client is unable to verify the certificate, it will abort the connection. This can happen if a client is re-routed to a phishing website.
4. To avoid requiring both the client and server to expend computational power to create their own private keys, the client sends a message encrypted with the server's public key. In the encrypted transmission is a "__pre-master key__", which is a shared secret key between the client and the server. The pre-master key is transferred under the server's public key encryption.
    - This would be a risk without digital certificate verification, as you would essentially be sending the wrong person the method of decrypting the information. __The digital certificate is what makes TLS possible.__
5. The server decrypts the message with its own private key, thus decrypting the pre-master key in that message. The decrypted pre-master key is the _master key_.
    - Sometimes, part of step 5 is the server first verifying the client's digital certificate, if required by the server.
6. The server sends a confirmation to the client encrypted in the shared master key. The client decrypts the confirmation message with the master key decryption.
    - Both the client and the server have saved computational power by engaging in symmetric encryption from this point on.
7. The remainder of the transmissions is done by both the client and server encrypting and decrypting the data with the master key. 
    - No private data is ever transferred until step 6 is complete.

# E. World Wide Web and HTTP & HTML

HTTP is a uniform protocol for determining what to do with the data packets that are transferred between a client and a web server (does the client want ("GET") information, or is it sending the server ("POST") information?). Because HTTP is uniformly used across the web, "HT"ML is uniformly used as the basis for viewing websites.


An ISP can see every HTTP and HTTPS request that goes over their network.

***

## E.1. Hypertext Transfer Protocol (HTTP) Step-by-Step

The following is done via HTTP everytime a URL beginning with "http://" is used:
1. A web browser uses a DNS resolver to map the desired URL with an IP address.
2. The browser sends an HTTP request to the destination IP address.
    - An example is "GET" /index.html HTTP/1.1 Host: www. example.com for a request to visit the website www. example.com/ index.html.
        - Shows (1) what to get, (2) under what protocol it should be got, and (3) where to get it from.
3. Host computer sends back requested data along with corresponding metadata, such as an HTTP status code ("200" is okay, "404" is not found, "403" is forbidden, etc.).
    - The requested data is the website's HTML code.
4. The client's web browser renders the HTML code to represent the website to the client.

***

## E.2. HTTPS vs. HTTP

HTTPS is "HyperText Transfer Protocol Secure", which encrypts HTTP requests and responses with the TLS protocol. 
- Also known as "HTTP over TLS".
- URLs have "https://" instead of "http://"
- HTTPS will prevent a connection to a website if the website's digital certificate is not valid.


With the additional layer of security, third-parties are not able to see the content of the communications between the client and server, though they can still see the IP addresses associated with the connection, the duration of the connection, and other metadata.

# F. Malware & Cybercrime Definitions

### Code-Based Cybercrime

- __Malware:__ Malicious software unknowingly installed on a computer. 
    - Types of Malware:
        - __Trojan Horse:__ A program that looks like a legitimate program but is not. It causes damage when the user runs the program.
        - __Virus:__ Damaging code inside an otherwise legitimate program. It replicates its code into other files on the system.
        - __Worm:__ A type of virus that replicates to different computers within a network, often hogging bandwidth.
    - Examples of Malware:
        - __Spyware:__ Steals data and sends it to the malware creators.
            - One example is _keylogging_, where the malware creators record every keystroke of a user. Often done in conjunction with finding passwords.
        - __Adware:__ Pop-ups and the like, which typically generate revenue for the malware creators.
        - __Ransomware:__ Encrypting a user's data or applications, demanding the user pay a ransom to the malware creators for access.
        - __Cryptomining Malware:__ Utilizes a computer's resources to mine for cryptocurrency for the benefit of the malware creators.
        
       
- __Botnets:__ Collection of computers all run by the same bot, to launch a large-scale cyber attack. Examples of Botnet applications:
    - Largescale malware distribution.
    - __Distributed Denial-of-Service (DDoS) Attacks:__ Forcing computers to send a wave of HTTP requests to a specific server with the aim of crashing it (or at least significantly slowing it down).
    
### Human-Based Cybercrime

- __Phishing:__ Tricking a user into divulging his/her private information, often done by looking like a legitimate website (banking website, online store, etc.), which is called _DNS Spoofing_.
- __Spear-Phishing:__ A type of phishing where phishing emails are targeted at a specific group of individuals, such as the employees of a company to obtain the company's data.

## Functions

Parameters of Python functions:

    - for a function `def func(a=1, b=2, c=3): pass`: 
        - a, b, and c are the default parameter/argument values. If you did func(), it would readily assume the default values of 1, 2, and 3, respectively.
        - Conversely, keyword arguments are _called_ - they are not the default. This allows for out-of-order calls too, like func(c=100, a=12, b=20). If "b" wasn't passed as an argument, again, it would assume the default value of b=2.
- Breaks in a function are used to stop and exit from executing _only_ the enclosed loop in which the "break" resides.
    
### Global vs. Local Differences

- __Global__ means outside of a function, i.e., applies globally. __Local__ means defined and used only in a specific context, such as only within a function.
- For example, "global built-in functions" are list, sum, and other basic Python built-in functionality that can be used both inside and outside functions.
    - Global built-in functions _can_ be redefined (e.g., `list = 'alfie'` making `list(0,1,2,3)` broken), but that is ill-advised.
    

- The __LEGB Rule__ identifies the order in which Python searches a variable for use:
    - __(L)__ _Local_: Inside a function or class. If the variable is not found there, it will look to 
    - __(E)__ _Enclosed_: Outside portions of the function (not local to the specific indented level of the function).
    - __(G)__ _Global_: If not in a function, will look to global variables. That's why list = alfie changes list before
    - __(B)__ _Built-in_: If not defined globally, will look to built-in functionality for variable's purpose.
    <br><br>
    - Storing local variables are __faster__ to lookup than global variables. _Why?_ The bytecode uses a `STORE_FAST` action instead of the global variable's `STORE_NAME` action. Why this is the case just comes down to convention (based on my research).
